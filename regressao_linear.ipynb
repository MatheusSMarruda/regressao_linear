```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

#input dos dados do testes
nome_graph = input("Informe o nome do grafico")
qnt_teste = int(input(("Informe quantos testes foram feitos")))
i_corrente = []  # Coordenadas x que você escolheu
u_tensao = []  # Coordenadas y que você escolheu

for i in range(qnt_teste):
    teste_tensao = float(input("Digite o valor da tensão informado na fonte"))
    u_tensao.append(teste_tensao)
    teste_corrente = float(input("Digite o valor da tensão informado no voltimetro"))
    i_corrente.append(teste_corrente)

# Gráfico de dispersão
plt.scatter(i_corrente, u_tensao, color='blue', label='Dados')

# Convertendo as listas para arrays numpy
i_corrente_np = np.array(i_corrente)
u_tensao_np = np.array(u_tensao)

# Cálculo da linha de tendência (regressão linear)
coeficientes = np.polyfit(i_corrente_np, u_tensao_np, 1)  # 1 indica uma reta
linha_tendencia = np.polyval(coeficientes, i_corrente_np)

# Plot da linha de tendência
plt.plot(i_corrente_np, linha_tendencia, color='red', label='Linha de Tendência')

# Exibir o gráfico
plt.xlabel("Corrente (I)")  # Rótulo do eixo x
plt.ylabel("Tensão (V)")    # Rótulo do eixo y
plt.title(nome_graph)  # Título do gráfico
plt.show()
print(i_corrente_np)
print(u_tensao_np)
```
